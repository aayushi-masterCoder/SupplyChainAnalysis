What is Supply Chain Management: handling of entire production flow of good or services, starting from row components all the way to delivering final product to consumer.
https://www.ibm.com/topics/supply-chain-management

Columns description in dataset:
* **Product Type**
* **SKU**: Stock Keeping Unit (barcode of the product)
* **Price**: price of product
* **Availability** Available product
* **Number of products sold**: sold products
* **Revenue generated**: total ammount of income generated by the sale of product,service,goods,etc..
* **Customer demographics**: Customer identity or individuality
* **Stock levels**: the amount of stock available througout the distribution network
* **Order quantities**
* **Shipping times**
* **Shipping carriers**
* **Shipping costs**
* **Supplier name**
* **Location**
* **Lead time**: time it takes from purchase order gets create with supplier untill order gets delivered to customer.
* **Production volumes**: the maximum product output a company can produce using its available resources over a specified amount of time
* **Manufacturing lead time**: the time it takes to produce a product or service
* **Manufacturing costs**: the costs incurred during the production of a product.
* **Inspection results**: 
* **Defect rates**: the portion of defective elements in relation to all items produced
* **Transportation modes**
* **Routes**
* **Costs**
